#### Типы рекламы в мобильных играх

1. **Rewarded Video Ads**  
This ad format is equally beloved by players, developers, and advertisers.

After players watch rewarded video ads – they get a reward in return. At the same time, the developer gets ad revenue, while the advertisers get viewability and high CTRs.

It’s a win-win-win situation.

These rewards can be any kind of in-game feature: currency, lives, points, gems, etc. All of these are the things they would normally have to earn or purchase. With rewarded video ads, the only currency they have to use is their time and attention.

2. **Playable Ads**   
Playable ads are also one of the user’s favorites.

As the name suggests, this ad type allows the users to play. Playable ads are mini-games that appear during a game session.

They allow the user to engage with and try out another game without installing it. After the users complete the micro-game, they see a call to action (e.g. install now).

This ad format results in a positive user experience and high eCPMs. Players perceive them as the most enjoyable out of all ad types. In fact, they perform up to 8 times better than interstitial ads (SmartyAds). This benefits the developer because they don’t disrupt the user gameplay experience.

3. **Interstitial Ads**  
Interstitial ads are a full-screen ad format that can be skipped after users watch them for a while.

Taking over users’ screens, interstitial ads can be either static or dynamic (interstitial video ads). For better engagement rates, it is recommended to utilize interactive content.

When it comes to eCPMs in interstitial ads, they average at $4 to $6 but can climb to as high as $20.

The players don’t enjoy this ad format as much as the first two.

However, there are several factors that play a role in how much revenue these ads will generate. The game genre, platform, location, and most importantly – how and when you display the ads.

The thing is, interstitial ads can be disruptive and annoying for players.

The golden rule to prevent this from happening is placing interstitial ads in natural gameplay pauses.

Here are some additional tips for making interstitial ads unobtrusive and effective:

use predictable placements
keep an eye on frequency
pay attention to other ad formats

4. **Native Banner Ads**  
This ad format is an improved version of a well-known classic – banner ads. Native banner ads are blended into gameplay so they feel like a part of it.

If this ad is well-made and strategically placed, it shouldn’t even seem like an ad. They should be discrete and never interrupt gameplay. Banners are a good option for games that come in a portrait format, as there is a smaller chance they will interfere with the gameplay.

Advertisers use them mainly because they are cheaper and simpler to produce than other ad formats. As for developers, in terms of advertising revenue, they are far behind rewarded ads, playable ads, and interstitials.

However, used in a combination with other ad formats, they can make an effective monetization strategy.

5. **Offerwalls**  
This ad format works on a similar principle as rewarded video ads.

When users stumble upon an offerwall, they have to complete some kind of action to receive a reward.

An offerwall looks like a mini shop inside the game, listing the actions for the users to take part in.

Here are examples of possible actions that can appear on an offerwall:

completing a survey
reaching level 10 in a game
making an in-app purchase
playing a mini-game
downloading a game
The biggest benefits developers can gain from this ad type are longer play sessions and high eCMPs.

Offerwall eCMPs are much higher than for other ad formats. Especially in certain genres.

For Android users in the U.S.,  eCPMs for mid-core games average at $980. They are the highest in the RPG genre where the average eCMPs are $1,670 (ironSource).

On the other hand, the eCPMs in the casual category average at $300.

Let me remind you – the second-best are rewarded video ads. Their eCPMs “only” go as high as $50.

If you decide to include offerwall ads into your game, make sure they are compatible with the players’ interests. Also, note that they work better in games that have a currency as an in-game feature.

```python
# Функция для вывода пропущенных значений в датафрейме, просто вызовите ее, указав нужный датафрейм в параметрах
def missing_values_tab(df):
    #Подсчет количества пропущенных значений
    mis_val = df.isnull().sum()
    #Подсчет доли пропущенных значений в колонке в процентах
    mis_val_percent = round(100 * df.isnull().sum() / len(df),2)
    #Объединяем две созданных колонки в один датафрейм с параметром axis=1, чтобы добавить их справа-слева
    mv_table = pd.concat([mis_val, mis_val_percent], axis=1)  
    #Переименование колонок
    mv_table = mv_table.rename(columns = {0 : 'Пропущенные значения', 1 : '% от всех значений'})
    #Создаем колонку с типом данных 
    mv_table['Тип данных'] = df.dtypes
    mv_table = mv_table[mv_table.iloc[:,1]!=0].sort_values(by='Пропущенные значения',ascending=False)
    print ("Ваш датафрейм содержит " + str(df.shape[1]) + " колонок и " + str(df.shape[0]) + " строк.\n")
    print("Он имеет  " + str(mv_table.shape[0]) +" колонок с пропущенными значениями.")    
    return display(mv_table) 
```

```python
def get_profiles(sessions, user_source, ad_costs,events=None, event_names=[],sort=False):

    # находим параметры первых посещений
    sessions = sessions.merge(user_source, on = ['user_id'], how = 'left')
    profiles = (
        sessions.sort_values(by=['user_id', 'event_datetime'])
        .groupby('user_id')
        .agg(
            {
                'event_datetime': 'first',                
                'event': 'first',
                'building_type': 'first',
                'project_type': 'first',
                'channel':'first'
            }
        )
        .rename(columns={'event_datetime': 'first_event_datetime'})
        .reset_index()
    )

    # для когортного анализа определяем дату первого посещения
    # и первый день месяца, в который это посещение произошло
    profiles['first_event_date'] = profiles['first_event_datetime'].astype('datetime64[D]') 
    

    # добавляем признак пользователя из канала
    profiles['acquired_user'] = profiles['user_id'].isin(user_source['user_id'].unique())
    
    # добавляем флаги для всех событий из event_names
    for event in event_names:
        if event in events['event_name'].unique():
            profiles[event] = profiles['user_id'].isin(
                events.query('event_name == @event')['user_id'].unique()
            )

    new_users = (
        profiles.groupby(['first_event_date','channel'])
        .agg({'user_id': 'nunique'})
        .rename(columns={'user_id': 'unique_users'})
        .reset_index()
    )   
    #Так как затраты на рекламу идут с лагом в один день, то надо его прибавить 
    ad_costs['date'] = ad_costs['date'] +timedelta(days= 1)
     # объединяем траты на рекламу и число привлечённых пользователей
    ad_costs = ad_costs.merge(new_users, left_on=['date','channel'],
                              right_on = ['first_event_date','channel'], 
                              how='left')
     # делим рекламные расходы на число привлечённых пользователей
    ad_costs['acquisition_cost'] = ad_costs['cost'] / ad_costs['unique_users']
   
    # добавляем стоимость привлечения в профили
    profiles = profiles.merge(
        ad_costs[['date','channel', 'cost','unique_users','acquisition_cost',]],
         right_on=['date','channel'], left_on = ['first_event_date','channel'],
        how='left')
    
    # стоимость привлечения органических пользователей равна нулю
    profiles['acquisition_cost'] = profiles['acquisition_cost'].fillna(0)
    profiles = profiles.drop(columns = ['date','project_type'], axis = 1)
    if sort:
    # Сортировка колонок
        cols = profiles.columns.to_list()    
        cols = cols[:1] + cols[-5:-4]+cols[1:5]+cols[-4:]
        profiles = profiles[cols]
    return profiles, ad_costs
```

```python
''' Создание списков колонок по типу данных'''
def create_column_lists(data):
    #Список типов, принимаемых за численные
    numerics_types = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
    #Создание списка числовых колонок    
    numeric_cols = [name for name in data.columns if data[name].dtype in numerics_types]
    #Список непрерывных числовых колонок, если больше 25 уникальных значений
    cont_cols = [name for name in numeric_cols if len(data[name].unique()) >  25]
    #Список дискретных числовых колонок, если меньше или равно 25 уникальных значений
    disc_cols = [name for name in numeric_cols if len(data[name].unique()) <= 25]
    #Список колонок с категориальными значениями
    categorical_cols = [name for name in data.columns if data[name].dtype == 'object' and len(data[name].unique()) < 10
               or data[name].dtype == 'bool']    
    result_list = [numeric_cols, cont_cols, disc_cols, categorical_cols]
    return result_list
```

```python
def show_values (ax, orient = 'bar'):
    '''Функция для показа значений для вертикальных или горизонтальных бар-графиков'''
    rects = ax.patches
    if orient =='bar':
        
        for rect in rects:    
            height = np.nan_to_num(rect.get_height(), 0)
            if height != 0:
                ax.text(rect.get_x() + rect.get_width() / 2, height, round(height,1),
                ha="center", va="bottom", weight='medium',size='small', style='oblique')  
    if orient =='barh':        
        for bar in rects:    
            width = np.nan_to_num(bar.get_width(), 0)    
            if width != 0:
                ax.text(width, bar.get_y()+0.2, str(round(width, 2)),fontsize='large', style='oblique',
                weight='medium', color='k',ha='left', size = 'small') 
```

```python
def draw_barchart(data,                # таблица данных
                  column,              # колонка таблицы
                  figsize = (10,5),    # размер графика
                  show_numbers = False, #показывать цифры над столбиками или справа от них
                  kind = 'bar',         # выбор графика bar или barh             
                  title = None,         # название графика
                  legend=False,         # показывать легенду
                  grid = False,         # показывать сетку
                  color = '#1f77b4',    # выбор цвета
                 ax = None,             # задать ось
                 xlabel = '',           # показывать название оси x
                  ylabel = '',          # показывать название оси y
                  rot = 0):             # угол наклона значений осей 
    '''Функция для рисования горизонтального и вертикального бар-графика'''
    ax = data[column].plot(
                kind=kind, edgecolor='black',  grid=grid, figsize = figsize,
                title = f'{column}', 
                xlabel = xlabel, 
                ylabel = ylabel, 
                legend = legend,
                color = color, ax = ax,rot = rot )
            #Убирание рамок вложенных графиков
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)                  
    if show_numbers:
            show_values (ax, orient = kind)  
```

```python
def draw_barplots(data,                  #таблица данных                             
                  column_list=None,      # список колонок для показа, по умолчанию все колонки, тип list
                  maxcolumns = 1,        #задать ограничение по числу колонок
                  kind = 'bar',         # bar или barh 
                  height_factor = 1,    #скалирование по высоте
                  width_factor = 1,     # скалирование по ширине                  
                  figsize = None,       # принудительное указание размера (x,y)
                  show_numbers = False, # показывать значения над столбиками
                  xlabels_rot = 45,     # наклон надписей отметок на оси x,y
                  format_dates = False, # форматировать даты на осях, если мешают часы, минуты секунды                  
                  xlabels = '',         # показывать название оси х
                  ylabels = '',         # показывать название оси y
                  main_title = '',       # главное название
                  grid = False,         # сетка
                             
                  color = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22', '#17becf']
                 ):
    '''Функция для рисования горизонтальных и вертикальных бар-графиков с динамическим размером'''
    def dates_axis_ticklabels_formatter (data):
        '''Функция для форматирования дат'''
        return [x.strftime("%m-%d") for x in data.index]

       
    if column_list:
        columns =column_list
    else:
        columns =data.columns 
    
    ncols = min(len(columns),maxcolumns)
    nrows = math.ceil(len(columns)/ncols)     
    
    if main_title:
            fig.suptitle(main_title)
    if not figsize:
        fig, axes = plt.subplots(nrows=nrows, ncols=ncols, 
                                 figsize=(ncols*4*width_factor, nrows*4*height_factor))
    else:
        fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize = figsize)
    for i in range(len(columns)):
            draw_barchart(data = data, column = columns[i],
                          kind = kind, 
                          title = f'{columns[i]}',
                          color = color[i],
                          ax = plt.subplot(nrows,ncols, i+1),                           
                          figsize = (ncols*4*width_factor, nrows*4*height_factor),
                          show_numbers = show_numbers,                          
                          grid = grid,
                          xlabel = xlabels,
                          ylabel = ylabels, rot = xlabels_rot)                       
            if format_dates and kind == 'bar':
                plt.subplot(nrows,ncols, i+1).set_xticklabels(dates_axis_ticklabels_formatter(data), rotation=xlabels_rot )
            if format_dates and kind == 'barh':
                plt.subplot(nrows,ncols, i+1).set_yticklabels(dates_axis_ticklabels_formatter(data), rotation=0 )
            if show_numbers:
                 show_values(ax, orient = kind)         
            i+=1 
    #Убрать ненужные оси   
    unused_axes = (nrows*ncols)-len(columns)
    for unused_axes in range(unused_axes):
            axes[nrows-1, -1-unused_axes].remove()        
    plt.subplots_adjust(hspace=0.3*(1/height_factor),wspace=0.1*1/width_factor)
    plt.tight_layout()
    plt.show()
```

```python
# функция для расчёта LTV и ROI

def get_ltv(
    profiles,    
    observation_date,
    horizon_days,
    dimensions=[],
    ignore_horizon=False,
):

    # исключаем пользователей, не «доживших» до горизонта анализа
    last_suitable_acquisition_date = observation_date
    if not ignore_horizon:
        last_suitable_acquisition_date = observation_date - timedelta(
            days=horizon_days - 1
        )
    result_raw = profiles.query('first_event_date <= @last_suitable_acquisition_date')     
    
    # группируем по cohort, если в dimensions ничего нет
    if len(dimensions) == 0:
        result_raw['cohort'] = 'All users'
        dimensions = dimensions + ['cohort']

    # функция группировки по желаемым признакам
    def group_by_dimensions(df, dims, horizon_days):
        # строим «треугольную» таблицу выручки
        result = df.pivot_table(
            index=dims, columns='lifetime', values='revenue', aggfunc='sum'
        )
        # находим сумму выручки с накоплением
        result = result.fillna(0).cumsum(axis=1)
        # вычисляем размеры когорт
        cohort_sizes = (
            df.groupby(dims)
            .agg({'user_id': 'nunique'})
            .rename(columns={'user_id': 'cohort_size'})
        )
        # объединяем размеры когорт и таблицу выручки
        result = cohort_sizes.merge(result, on=dims, how='left').fillna(0)
        # считаем LTV: делим каждую «ячейку» в строке на размер когорты
        result = result.div(result['cohort_size'], axis=0)
        # исключаем все лайфтаймы, превышающие горизонт анализа
        result = result[['cohort_size'] + list(range(horizon_days))]
        # восстанавливаем размеры когорт
        result['cohort_size'] = cohort_sizes

        # собираем датафрейм с данными пользователей и значениями CAC, 
        # добавляя параметры из dimensions
        cac = df[['user_id', 'acquisition_cost'] + dims].drop_duplicates()

        # считаем средний CAC по параметрам из dimensions
        cac = (
            cac.groupby(dims)
            .agg({'acquisition_cost': 'mean'})            
        )

        # считаем ROI: делим LTV на CAC
        roi = result.div(cac['acquisition_cost'], axis=0)

        # удаляем строки с бесконечным ROI
        roi = roi[~roi['cohort_size'].isin([np.inf])]

        # восстанавливаем размеры когорт в таблице ROI
        roi['cohort_size'] = cohort_sizes

        # добавляем CAC в таблицу ROI
        roi['acquisition_cost'] = cac['acquisition_cost']

        # в финальной таблице оставляем размеры когорт, CAC
        # и ROI в лайфтаймы, не превышающие горизонт анализа
        roi = roi[['cohort_size', 'acquisition_cost'] + list(range(horizon_days))]

        # возвращаем таблицы LTV и ROI
        return result, roi

    # получаем таблицы LTV и ROI
    result_grouped, roi_grouped = group_by_dimensions(
        result_raw, dimensions, horizon_days
    )

    # для таблиц динамики убираем 'cohort' из dimensions
    if 'cohort' in dimensions:
        dimensions = []

    # получаем таблицы динамики LTV и ROI
    result_in_time, roi_in_time = group_by_dimensions(
        result_raw, dimensions + ['first_event_date'], horizon_days
    )

    return (
        result_raw,  # сырые данные
        result_grouped,  # таблица LTV
        result_in_time,  # таблица динамики LTV
        roi_grouped,  # таблица ROI
        roi_in_time,  # таблица динамики ROI
    )
```

```python
def get_retention(
    data,                 # таблица данных
    observation_date,     # дата наблюдения
    horizon_days,         # горизонт планирования
    dimensions=[],        # список параметров для группировки
    ignore_horizon=False, # игнорировать горизонт и считать по всем данным
):

    # исключаем пользователей, не «доживших» до горизонта анализа
    last_suitable_acquisition_date = observation_date
    if not ignore_horizon:
        last_suitable_acquisition_date = observation_date - timedelta(
            days=horizon_days - 1
        )
     #Запрос из таблицы данных пользователей, первая дата которых меньше или равна предельно допустимой дате   
    result_raw = data.query('first_event_date <= @last_suitable_acquisition_date')
   
     #группируем по всем пользователям, если в dimensions ничего нет
    if len(dimensions) == 0:
        result_raw['cohort'] = 'All users'
        dimensions = dimensions + ['cohort']
    # функция для группировки таблицы по желаемым признакам
    def group_by_dimensions(df, dims, horizon_days):
        result = df.pivot_table(
            index=dims, columns='lifetime', values='user_id', aggfunc='nunique'
        )
        cohort_sizes = (
            df.groupby(dims)
            .agg({'user_id': 'nunique'})
            .rename(columns={'user_id': 'cohort_size'})
        )
        result = cohort_sizes.merge(result, on=dims, how='left').fillna(0)
        result = result.div(result['cohort_size'], axis=0)
        result = result[['cohort_size'] + list(range(horizon_days))]
        result['cohort_size'] = cohort_sizes
        return result

    # получаем таблицу удержания
    result_grouped = group_by_dimensions(result_raw, dimensions, horizon_days)

    # получаем таблицу динамики удержания
    result_in_time = group_by_dimensions(
        result_raw, dimensions + ['first_event_date'], horizon_days
    )

    # возвращаем обе таблицы и сырые данные
    return result_raw, result_grouped, result_in_time
```

```python
def filter_data(df, window):
    # для каждого столбца применяем скользящее среднее
    for column in df.columns.values:
        df[column] = df[column].rolling(window).mean() 
    return df
```

```python
#Функция для определения границ разброса значений у boxplots
def get_swipe_range(dataset):
    
    median = pd.DataFrame.median(dataset)
    mean = pd.DataFrame.mean(dataset)
    min_value =  pd.DataFrame.min(dataset)
    max_value = pd.DataFrame.max(dataset)
    quartile_1 = pd.DataFrame(dataset).quantile(0.25,numeric_only=False)
    quartile_3 = pd.DataFrame(dataset).quantile(0.75,numeric_only=False)
    # Interquartile range
    iqr = round(quartile_3-quartile_1,2)
    #Вычисление границ размаха
    left_swipe_value = round(quartile_1-1.5*iqr,2)
    right_swipe_value = round(quartile_3+1.5*iqr,2)    
    print('Границы разброса значений для колонки:',dataset.name)         
    print('Левое значение: %s' % left_swipe_value.to_string())
    print('Правое значение: %s' % right_swipe_value.to_string())
```

#### Создание сводной таблицы

Создание единой таблицы на основе profiles и game_actions без колонки project_type, так как она дублирует информацию в колонке event

```python
main_table = profiles.merge(game_actions[['event','event_datetime','user_id','building_type']], 
                            on = ['user_id'], 
                            how = 'right').rename(columns = {'event_x':'first_event_type',
                                                             'building_type_x':'first_event_building_type',
                                                             'event_y':'event',
                                                             'building_type_y':'building_type',
                                                             'cost':'channel_costs_per_cohort'
                                                            })
numeric_cols, cont_cols, disc_cols, categorical_cols = create_column_lists(main_table)
```

```python
main_table.name = 'Table main_table'
#Создание колонки с количеством дней, прошедших от первого события игрока до текущего (lifetime):
main_table['lifetime'] = (main_table['event_datetime'] - main_table['first_event_datetime']).dt.days
#Создание еще одной колонки с датами, так как предыдущая стала первой датой
main_table['event_date'] = main_table['event_datetime'].astype('datetime64[D]') 
#Колонка с датой последнего посещения игры
main_table['last_event_date'] = main_table.groupby('user_id')['event_date'].transform(max)
# Продолжительность игровой жизни игрока в днях
main_table['total_lifetime'] = main_table.groupby('user_id')['lifetime'].transform(max)
#Создание колонки с номером недели
main_table['week'] = main_table['event_date'].dt.week
#Колонка с датой,временем последнего посещения игры 
main_table['last_event_datetime'] = main_table.sort_values(['user_id','event_datetime']).groupby('user_id')['event_datetime'].transform(max)
#Колонка с длительностью игры в часах
main_table['duration_hrs'] = ((main_table['last_event_datetime'] - main_table['first_event_datetime'])/ pd.Timedelta(hours=1) ).round(2)
```

Создание колонки категорий игроков по результатам первого уровня:
 
 - выборка уникальных  user_id игроков с  `event=finished_stage_1`, это игроки, прошедшие первый уровень;
 - следовательно, остальные user_id имеют игроки, не прошедшие первый уровень; 
 - у таких игроков пропуски в этой колонке будут заполнены значением `not_completed`;
 - игрокам, у которых есть событие `event==project`, будет присвоена категория `completed_project`
 - в остальных случаях - значение `defeated_enemy`.  

```python
#Таблица игроков, прошедших уровень
level_completed_players = main_table.query('event != "building"')
project_finishers = level_completed_players.query('event =="project"')['user_id'].unique()
#Присваивание категорий
level_completed_players.loc[~level_completed_players['user_id'].isin(project_finishers),'player_result'] = 'defeated_enemy'
level_completed_players.loc[level_completed_players['event']=='project','player_result'] = 'completed_project'
#Список игроков, закончивших уровень через победу над противником
id_list_pvpers = level_completed_players.query('player_result=="defeated_enemy"')['user_id'].unique()
```

```python
main_table.loc[main_table['user_id'].isin(project_finishers),'player_result'] ='completed_project'
main_table.loc[main_table['user_id'].isin(id_list_pvpers),'player_result'] ='defeated_enemy'   
main_table['player_result'] = main_table['player_result'].fillna('not_completed')
```

```python
# Создание колонки с категориями игроков завершивших уровень и нет
main_table = main_table.assign(
    level_completed = lambda row: row['player_result'].map(
        {'defeated_enemy': 'yes', 
         'completed_project': 'yes', 
         'not_completed': 'no'        
        }
    ))
```

### Гипотеза 1. О различии среднего времени прохождения <a class="anchor" id="id_5a"></a>

*Нулевая гипотеза*: среднее время прохождения уровня у пользователей, котороые заканчивают уровень проектом, равно среднему времени прохождения уровня у игроков, которые заканчивают уровень победой над другим игроком.  
*Альтернативная гипотеза*: среднее время прохождения уровня у вышеупомянутых категорий игроков не равно.  
Проверить эти гипотезы можно с помощью t-тестов, но при соблюдении следующих условий:  
 - Генеральные совокупности не должны зависеть друг от друга.
 - Выборочные средние должны быть нормально распределены.
   А если выборка всего одна, то её среднее должно иметь нормальное распределение для разных выборок этого размера из определённой генеральной совокупности.
 - Дисперсии рассматриваемых генеральных совокупностей должны быть равны.
